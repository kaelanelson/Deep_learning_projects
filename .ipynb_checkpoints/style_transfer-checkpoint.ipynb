**Goal:** To generate a stylistic filter on top of an image. For example, extracting the swirls from Van Gogh's "A Starry Night" and applying it to an image of buildings.

```python
class VGGIntermediate(nn.Module):
  def __init__(self, requested=[]):
    super(VGGIntermediate, self).__init__()
    
    self.intermediates = {}
    self.vgg = models.vgg16(pretrained=True).features.eval()
    for i, m in enumerate(self.vgg.children()):
        if isinstance(m, nn.ReLU):   # we want to set the relu layers to NOT do the relu in place. 
          m.inplace = False          # the model has a hard time going backwards on the in place functions. 
          
        if i in requested:
          def curry(i):
            def hook(module, input, output):
              self.intermediates[i] = output
            return hook
          m.register_forward_hook(curry(i))
    
  def forward(self, x):
    self.vgg(x)  
    return self.intermediates
```

# Get content and style activation layers

```python
def get_models(requested_vals, content_img, style_img):
    # get content layers
    vgg = VGGIntermediate(requested=requested_vals.values())
    vgg.cuda()
    content_layer = vgg(content_img.cuda()) # returns a model

    # get style layers
    vgg = VGGIntermediate(requested=requested_vals.values())
    vgg.cuda()
    style_layer = vgg(style_img.cuda())

    return content_layer, style_layer
```

```python
def get_requested_layers(requested_vals, content_layer, style_layer):
  # gives you the specific layers used in the paper
  content_output_activation = content_layer[requested_vals["conv4_2"]] 
  style_output_activation = [style_layer[requested_vals[i]] for i in ["conv1_1", "conv2_1", "conv3_1", "conv4_1", "conv5_1"]]
  return content_output_activation, style_output_activation
```

# Content Loss

```python
class Content_Loss(nn.Module):
  def __init__(self, target_image):
    super(Content_Loss, self).__init__()
    self.target_image = target_image.detach()
  
  def forward(self, input_image):
    self.loss = F.mse_loss(input_image, self.target_image)
    return self.loss
```

# Style Loss

```python
def compute_gram(input_image):
  a, b, c, d = input_image.shape
  M = c*d # M - number of pixels
  N = a*b # a = 1, number of channels/kernels
  new_im = input_image.view(N, M) # resize the image to be num kernels x num pixels
  G = torch.mm(new_im, new_im.t())
  return G.div(M*N)

class Style_Loss(nn.Module):
  def __init__(self, original_im):
    super(Style_Loss, self).__init__()
    self.target_image = compute_gram(original_im).detach()
    
  def forward(self, input_image):
    G = compute_gram(input_image)
    self.loss = F.mse_loss(G, self.target_image)
    return self.loss
```

# Optimization

```python
vgg_names = ["conv1_1", "relu1_1", "conv1_2", "relu1_2", "maxpool1", 
             "conv2_1", "relu2_1", "conv2_2", "relu2_2", "maxpool2",
             "conv3_1", "relu3_1", "conv3_2", "relu3_2", "conv3_3", "relu3_3","maxpool3",
             "conv4_1", "relu4_1", "conv4_2", "relu4_2", "conv4_3", "relu4_3","maxpool4",
             "conv5_1", "relu5_1", "conv5_2", "relu5_2", "conv5_3", "relu5_3","maxpool5"]
requested_vals = {name: i for i, name in enumerate(vgg_names)}
style_loss = []
content_loss = []

# grabs all layeres in the models
content_layers, style_layers = get_models(requested_vals, content_image, style_image)

# grabs desired layers to optimize on - might need the shape on target_hat
content_hat, style_hat = get_requested_layers(requested_vals, content_layers, style_layers)

# creates model object for each desired layer
style_loss_models = [Style_Loss(layer) for layer in style_hat]
content_loss_model = Content_Loss(content_hat)
```

```python
initializes input image layers as a clone of content image
input_image = content_image.clone()
vgg = VGGIntermediate(requested=requested_vals.values())
vgg.cuda()

# initialize optimizer
image_optimizer = optim.Adam([input_image.requires_grad_()], lr=.1)

style_weight = 100000
content_weight = 1

# Optimizing
for epoch in tqdm(range(1000)):  
  image_optimizer.zero_grad()
  
# get input image layers - initialized as a clone of content image
  input_layers = vgg(input_image.cuda())
  target_hat = [input_layers[requested_vals[i]] for i in ["conv1_1", "conv2_1", "conv3_1", "conv4_1", "conv5_1"]]
  
  # Compute Loss
  sl = 0
  for i in range(5):   
    sl += style_loss_models[i](target_hat[i])
  #)
  cl = content_loss_model(input_layers[requested_vals["conv4_2"]])
  
  total_loss = style_weight*sl + content_weight*cl
  total_loss.backward()

  image_optimizer.step()
  
```

```python
# inverse normalizes image
inverse_norm = transforms.Normalize(mean=[-.485/0.229, -0.465/0.224, -0.4068/0.255],
                                    std=[1/0.229, 1/0.224, 1/0.225])

def plot_img(tensor, title=None):
    image = tensor.cpu().clone()  # we clone the tensor to not do changes on it
    image = image.squeeze(0)      # remove the fake batch dimension
    image = inverse_norm(image)
    image.clamp_(0,1)
    plt.imshow(image.numpy().transpose(1,2, 0))
    plt.axis('off')
    
    if title is not None:
        plt.title(title)

plot_img(input_image.detach(), title='Generated Image')
```
